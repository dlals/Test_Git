```python
from sample import newborn as nb
from sample import year

py.figure()

py.plot(nb, label = 'baby count', linestyle= "--", marker= '.')
py.savefig('line1.png')

py.legend()
py.savefig('line2.png')

py.grid()
py.savefig('line3.png')

py.title('Newborn baby of year')
py.xlabel('year')
py.ylabel('newporn')
py.savefig('line4.png')

py.xticks([0, 1, 2, 3, 4], year)
py.savefig('line5.png')
py.show()
py.close()
```

```python
from sample import grade
keys = list(grade.keys())
values = list(grade.values())
xpos = list(range(0,len(values)))
```

```python
py.figure()

py.plot(values, label= 'grade', color= '#ff00ff')

py.legend()
py.grid()
py.title('Level Grade')
py.xlabel('level')
py.ylabel('point')
py.savefig('line3-1.png')

py.xticks(xpos, keys)
py.ylim(0, 100)
py.xlim(0,3)
py.savefig('lin3-2.png')
py.show()
py.close()
```
