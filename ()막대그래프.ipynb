```python
from matplotlib import pyplot
from sample import vote
from collections import Counter

cnt = Counter(vote)
cnt_dict = dict(cnt)

values = list(cnt_dict.values())
keys = list(cnt_dict.keys())

pyplot.rcParams["font.family"] = "HYHeadLine-Medium"
pyplot.rcParams["font.size"] = 12
pyplot.rcParams["figure.figsize"] = (12, 8)

pyplot.figure()

pyplot.bar(keys, values, label="득표수")
pyplot.grid()
pyplot.title("어벤져스 인기투표")
pyplot.xlabel("캐릭터")
pyplot.ylabel("득표수")
pyplot.legend()

for i ,v in enumerate(values):
    str_val = "%d표" % v
    pyplot.text(keys[i], v, str_val, fontsize=9, color='#ff6600',
               horizontalalignment='center', verticalalignment='bottom')


pyplot.show()
pyplot.close()
```

```python
from matplotlib import pyplot

from sample import newborn
x = list(range(0, len(newborn)))

pyplot.rcParams['font.family']='Hancom MalangMalang'
pyplot.rcParams['font.size']= 12
pyplot.rcParams['figure.figsize']=(12,8)
pyplot.figure()
pyplot.bar(x, newborn, label='신생아 수', color='#ecc2f9')
pyplot.title('2013~2017년도 신생아 수')
pyplot.xlabel('년도')
pyplot.ylabel('신생아 수')
pyplot.grid()
pyplot.xticks(x, [2013, 2014, 2015, 2016, 2017])

for i, v in enumerate(newborn):
    str_val="%d명" % v
    pyplot.text(x[i], v, str_val, fontsize=9, color='#000000',
               horizontalalignment='center', verticalalignment='bottom')
pyplot.ylim(300000, 450000)
pyplot.show()
pyplot.close()
```
